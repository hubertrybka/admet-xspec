```python
import pandas as pd
from src.data.tanimoto import TanimotoCalculator
from src.data.featurizer import EcfpFeaturizer

human_smiles = pd.read_csv("data/datasets/liver/human/property_T1.2/data.csv")['smiles'].tolist()
tc = TanimotoCalculator(smiles_list=human_smiles, featurizer=EcfpFeaturizer(radius=2, n_bits=2048))

# Load MAO-A IC50 dataset
rat_maoa_ic50 = pd.read_csv("data/datasets/liver/mouse/property_T1.2/mouse_all/data.csv")
results = tc.run_batch(queries=rat_maoa_ic50['smiles'].tolist())
rat_maoa_ic50['min_tanimoto'] = results['min_distance']
filtered_rat_maoa_ic50 = rat_maoa_ic50[rat_maoa_ic50['min_tanimoto'] > 0.1]
filtered_rat_maoa_ic50.to_csv("data/datasets/liver/mouse/property_T1.2/mouse_unique/data.csv", index=False)
```
