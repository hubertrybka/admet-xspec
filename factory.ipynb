```python
def get_config(datasets, task):
    return f"""
    include 'configs/processing_plans/split.gin'

    include 'configs/data_interface/data_interface.gin'
    include 'configs/splitters/scaffold_tanimoto_5p.gin'
    include 'configs/featurizers/ecfp.gin'

    ProcessingPipeline.data_interface = %data_interface
    ProcessingPipeline.splitter = %splitter
    ProcessingPipeline.featurizer = %featurizer
    ProcessingPipeline.task_setting = '{task}'

    ProcessingPipeline.datasets = ["{'", "'.join(datasets)}"]

    ProcessingPipeline.test_filtering_origin_dataset = '{datasets[0]}'
    ProcessingPipeline.filter_against_whole_dataset = True
    """
```

```python
import pathlib

dataset_combos = [["AChE_human_IC50"],
        ["AChE_human_IC50", "AChE_rat_IC50"],
        ["AChE_human_IC50", "AChE_mouse_IC50"],
        ["AChE_human_IC50", "AChE_rat_IC50", "AChE_mouse_IC50"],
        ["MAO-A_human_IC50"],
        ["MAO-A_human_IC50", "MAO-A_rat_IC50"],
        ["liver_human_T12"],
        ["liver_human_T12", "liver_rat_T12"],
        ["liver_human_T12", "liver_mouse_T12"],
        ["liver_human_T12", "liver_rat_T12", "liver_mouse_T12"]]

for task in ["regression", "classification"]:
    for combo in dataset_combos:
        config_str = get_config(datasets=combo, task=task)
        combo_name = abs(hash("_".join(combo)))%(10**5)
        save_path = pathlib.Path(f"{task}/{combo_name}.gin")
        with open(save_path, "w") as f:
            f.write(config_str)
```
