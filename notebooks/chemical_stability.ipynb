```python
import pandas as pd

# Load the dataset
PATH = '../generated_smiles.csv'
df = pd.read_csv(PATH).sample(1000, random_state=42)
```

```python
from src.stability.chemstable import DeepChemStableWrapper

# Initialize the DeepChemStable wrapper
dcs_wrapper = DeepChemStableWrapper()

# Predict stability scores using DeepChemStable
predicted_scores = dcs_wrapper.predict(df['smiles'].tolist())
predicted_scores.head()
```

```python
from IPython.display import Image
import pathlib
from ipywidgets import interact
import ipywidgets as widgets

unstable = predicted_scores[predicted_scores['is_stable'] == 0]

# list all files in the figures directory
figures_dir = pathlib.Path("../external/DeepChemStable/figures")
mapping = {int(str(x).split('_')[0].split('/')[-1]): x for x in figures_dir.glob('*.jpg')}

unstable_idcs = unstable.index.tolist()

@interact(img_idx=widgets.Dropdown(
    options=unstable_idcs,
    description='Image Index:',
    disabled=False,
    style={'description_width': 'initial'},
    continuous_update=False,
    layout=widgets.Layout(width='50%'),
))
def display_image(img_idx):
    """ Display an image from the DeepChemStable figures directory based on the index."""
    return Image(filename=mapping[img_idx], width=500)
```
