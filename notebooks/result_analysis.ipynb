```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

targets = ['8JLO', '8JSO', '8W8A', '8WCA', '8ZSV', '9JKQ']
dfs = [pd.read_csv(f'../results/taar1_mixed/{target}/{target}_CNS_BBBP_LogD.csv')[['molecule', 'BBBP_class_probability', 'BBBP_class', 'LogD_pred']] for target in targets]
for target, df in zip(targets, dfs):
    df['target'] = target
df = pd.concat(dfs, ignore_index=True)
```

```python
# plot the distribution of logD
fig, axes = plt.subplots(2, 3, figsize=(12, 8), sharey=True, constrained_layout=True)
sns.set_context('talk')
sns.set_style('white')
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
for ax, color, (target, group) in zip(axes.flatten(), colors, df.groupby('target')):
    sns.kdeplot(group['LogD_pred'], ax=ax, fill=True, alpha=0.5, color=color)
    ax.set_title(target)
    ax.set_xlabel('pred. LogD')
    # add vertical lines at 0 and 5
    ax.axvline(0, color='red', linestyle='--', linewidth=2)
    ax.axvline(5, color='red', linestyle='--', linewidth=2)
    # set xlim to -2 to 6
    ax.set_xlim(-1, 6)
    ax.set_xticks([0, 2.5, 5])
# save figure
plt.savefig('../results/taar1_mixed/logd_distribution.png', dpi=300)
```

```python
# Plot the distribution of BBBP class probabilities for each target

fig, axes = plt.subplots(2, 3, figsize=(12, 8), sharey=True, constrained_layout=True)
sns.set_context('talk')
sns.set_style('white')
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
for ax, color, (target, group) in zip(axes.flatten(), colors, df.groupby('target')):
    sns.kdeplot(group['BBBP_class_probability'], ax=ax, fill=True, alpha=0.5, color=color)
    ax.set_title(target)
    ax.set_xlabel('BBBP+ prob.')
    ax.set_ylabel('')
    ax.set_yticks([])
    # set xlim to 0 to 1
    ax.set_xlim(-0.1, 1.1)
    ax.set_xticks([0, 0.5, 1])
    # add vertical line at 0.5
    ax.axvline(0.5, color='red', linestyle='--', linewidth=3)

# save figure
plt.savefig('../results/taar1_mixed/bbbp_class_prob_distribution.png', dpi=300)
```
